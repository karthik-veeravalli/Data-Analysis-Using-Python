```python
df = sns.load_dataset("iris")
sns.jointplot(x = "p_length", y = "p_width",data = iris,knd=
             "reg")
plt.show()
```

```python
a= sns.jointplot("s_length","s_width",data= iris)
```

```python
a=sns.jointplot("s_width","p_length",data = iris)
```

```python
a=sns.jointplot("s_width","p_length",data =iris,kind='kde')
```

```python
g=(sns.jointplot("s_length","p_width", data =iris,color="k").plot_joint(sns.kdeplot,n_levels = 10))
```

```python
import numpy as np
x,y=np.random.rand(2,300)
g=(sns.jointplot(x,y,kind= "hex").set_axis_labels("x","y"))
```

```python
sns.set_style("ticks")
sns.pairplot(iris,hue = "spices",diag_kind= "kde",kind ="scatter",palette = "husl")
plt.show()
```

```python
g = sns.pairplot(iris,hue = "spices",markers = ["o","s","D"])
plt.show()
```

```python
sns.stripplot(x="spices",y="p_length",data=iris)
plt.show()
```
